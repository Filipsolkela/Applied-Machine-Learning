```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from matplotlib.pyplot import figure

plt.style.use('dark_background')
```

```python
iris = datasets.load_iris()

random_indeces = np.arange(iris.data.shape[0])
np.random.shuffle(random_indeces)

X_train = iris.data[random_indeces[0:100], 0:2]
Y_train = iris.target[random_indeces[0:100]]
X_test = iris.data[random_indeces[0:100], 0:2]
Y_test = iris.target[random_indeces[0:100]]
```

```python
clf_knn = KNeighborsClassifier(n_neighbors=1)
clf_knn.fit(X_train,Y_train)

clf_dt = DecisionTreeClassifier()
clf_dt.fit(X_train, Y_train)
```

```python
print("Accuracy (empirical) of the k-NN after training is %2.2f"% float(clf_knn.score(X_train, Y_train)*100.0))
print("Accuracy (generalization) of the k-NN after training is %2.2f"% float(clf_knn.score(X_test, Y_test)*100.0))
```

```python
print("Accuracy (empirical) of the k-NN after training is %2.2f"% float(clf_dt.score(X_train, Y_train)*100.0))
print("Accuracy (generalization) of the k-NN after training is %2.2f"% float(clf_dt.score(X_test, Y_test)*100.0))
```
